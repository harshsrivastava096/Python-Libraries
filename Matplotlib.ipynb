# **MATPLOTLIB**

Data Visualisation: It is a process of converting the data into pictorial or graphical representation to understand the data in a better manner

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Graphs --> line graph

---



> plot()-> plot a line graph

> **Attributes:**

---


* marker-> the point of y based on x

* markersize-> size of the marker

* markerfacecolor-> color within the marker

* markeredgecolor-> color surounding the marker/the edge of the marker

* linewidth-> defines how much the width of the line is

* linestyle-> define the style of the line-> chose anyone from this as you like-> ['solid', 'dashed', 'dotted', 'dashdot', 'none']

* xlabel() --> to label the x axis

* ylabel() --> to label the y axis

* title()  --> to give the title of the graph

* plt.annotate()-> the co-ordinates of the markers

* legend()-> defines what the graph consistes of

> **Attributes of Legend:**

---


  * title-> title of legend

  * facecolor-> the background colour of the lengend box

  * shadow-> will create som shadow effect on the box

  * loc-> where you want the legend box to be-> ['best','lower right','lower left','upper right','upper left','center','lower center','upper center']



```python
x = np.arange(1,11)
y = x**2

plt.plot(
    x,y,
    label="square of X values",
    marker="o",
    markersize=4,
    markerfacecolor="Yellow",
    markeredgecolor="red",
    linewidth=2,
    linestyle="dashed")

plt.xlabel(
    "Value from 1 to 10",
    color="blue",
    fontsize=15)

plt.ylabel(
    "Square of X",
    color="Green",
    fontsize=15)

plt.title(
    "Square",
    color="Red",
    fontsize=15)

for i in range(len(x)):
  plt.annotate(f'({x[i]},{y[i]})', (x[i],y[i]))

plt.legend(title="sample", facecolor="Pink", shadow=True, loc="upper left")
```

#Task: Create a line graph where:

> z = x**3

> the line should be dotted

> show the markers also

```python
x = np.arange(1,11)
z = x**3

plt.plot(x,z,
         marker="o",
         markersize=4,
         markerfacecolor="Yellow",
         markeredgecolor="red",
         linewidth=2,
         linestyle="dotted")

plt.xlabel("Value from 1 to 10", color="blue", fontsize=15)

plt.ylabel("Cube of X", color="Green", fontsize=15)

plt.title("Cube", color="Red", fontsize=15)

for i in range(len(x)):
  plt.annotate(f'({x[i]},{z[i]})', (x[i],z[i]))
```

#Task: Create a line graph where:

> z = x**3

> the line should be dotted

> show the markers

> Annotate

> A legend box and keep it at lower right



```python
x = np.arange(1,11)
z = x**3
plt.plot(x,z,
         label="Cube of X",
         marker="o",
         markersize=4,
         markerfacecolor="Yellow",
         markeredgecolor="red",
         linewidth=2,
         linestyle="dotted")

plt.xlabel(
    "Values of  X from 1 to 10",
    color="blue",
    fontsize=15,
    fontstyle="oblique")

plt.ylabel(
    "Value of Z from 1 to 1000",
    color="Green",
    fontsize=15,
    fontstyle="italic")

plt.title(
    "Cube Chart",
    color="Red",
    fontsize=15)

for i in range(len(x)):
  plt.annotate(f'({x[i]},{z[i]})', (x[i],z[i]))

plt.legend(title="Sample Data", facecolor="Pink", shadow=True, loc="lower right")
```
