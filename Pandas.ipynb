Task:
> 0-18 --> Younger age

> 19-45 --> Middle age

> > 46 --> Old Age

```python
a = []
for i in t['age']:
  if 0 < i < 18 :
    a.append('Younger age')
  elif 189< i <45:
      a.append('Middle age')
  else:
      a.append('Old age')
t['age_group'] = a
t
```

# Task:
> Create a fare_group column having 3 categories where:

* 1) 0-200 -> Low
* 2) 201-400 -> Medium
* 3) More than 400 -> High

```python
f=[]
for i in t['fare']:
  if i<=200:
    f.append('Low')
  elif 200<i<=400:
    f.append('Medium')
  else:
    f.append('High')
t['fare_group']=f
t
```
