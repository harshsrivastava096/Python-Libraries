```python
#lists consume more memory and time than array
#time-> library
#process_time -> class
from time import process_time
python_list=[i for i in range(10000)] #creation of list with range 0-9999
start_time=process_time() #stores the recorded start time
python_list=[i+2 for i in python_list] #element-wise addition in the list
end_time=process_time() #stores the recorded end time
print(end_time-start_time) #displays how much time it took to perform the element-wise operation
```

```python
#statistics
#mean, median, mode(no there in numpy), standard deviation, variance

arr=np.array([4,3,6,2,7,9,10,8,1])
#mean-> average(sum of the elements divided by number of elements)
np.mean(arr)
np.median(arr)
#for median the values are first sorted, if the number of values are odd, you will get an whole number
#if it is even, then a value will be in decimal
#but the datatype will be float
#(1,2,3,4,6,7,8,9,10)-> sorted array
#np.mode(arr)--> numpy does not have mode
np.var(arr)#-> how much data is spread/what the spread of the data is
np.std(arr)#--> how far a data is from the mean
```

```python
#Rounding
#np.ceil()-> rounding the figure upwards( towards +ve infinity)/rounding up the decimal to the next whole number
#np.floor()-> rounding the figure downwards(towards -ve infinity)/rounding the decimal to the previous whole number
#np.around()-> follows Banker's rounding
#Banker's rounding rules:
#1) If the decimal point is less than .5 then it will round up the value to the previous whole number
#2) If the decimal point is more than .5 then it will round up the value to the next whole number
#3) If the decimal point is exactly .5 then it will round up the value to the nearest even number
#3.5-> 3,4 -> 4
#4.5-> 4,5 -> 4
a=np.array([1,2,3.5,4.5])
np.ceil(a)
np.floor(a)
np.around(a)
# [1.,2.,4.,5.]
```
